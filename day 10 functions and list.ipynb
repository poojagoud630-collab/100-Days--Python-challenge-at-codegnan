```python
# positional arguments 
#subtraction of two numbers using functions
def sub (n1,n2):
    res = n1 - n2 
    return res 
a = int(input())  
b = int(input()) # a,b = 10,20
print(sub(a, b))
print(sub(b, a))
```

```python
# keyword arguments
# subtraction of two numbers using functions
def sub(n1,n2):
    res = n1 - n2 
    return res 
a = int(input())  
b = int(input())
print(sub(a, b))
print(sub(b, a))
```

# list properties

```python
## list properties
# list elements access by using index positions
lst = [10,2,4,4.534, 'hi','hello']
print(lst[2])
print(lst[-3])
print(lst[2:])
```

```python
# list mutable
lst = [1,2,3,40,'hi','hello']
lst[0] = 100
print(lst)
```

```python
# string are immutable
str="asdfgh"
str[0]= "R"
print(str)
```

# read list of numbers if numer is even add 0.5 and if the number is odd add one

```python
#read list of numbers if numer is even add 0.5 and if the number is odd add one
##lst=[1,2,3,4,5]
lst=list(map(int,input().split()))
for i in range(0,len(lst)):
    if lst[i]%2 :
        lst[i]=lst[i]+1
    else:
         lst[i]=lst[i]+0.5
print(lst)
```
