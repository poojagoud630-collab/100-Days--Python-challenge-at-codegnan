# basic calculartor using functions

```python
def add(a,b):
    r=a+b
    return r
def sub(a,b):
    s=a-b
    return s
def mul(a,b):
    t=a*b
    return t
def div(a,b):
    if b==0 :
        return "zero division not possible"
    u=a/b
    return u
if __name__ == "__main__" :
    print("welcome to small level calculator") 
r=add(20,10)
print(r)
s=sub(20,10)
print(s)
t=mul(20,10)
print(t)
u=div(20,10)
print(u)
```

```python
def addition(a,b):
    return a+b
def subtraction(a,b):   
    return a-b
def multiple(a,b):   
    return a*b
def division(a,b):
    if b==0:
        return "Zero division not possible"
    return a/b
if __name__ == "__main__":
    print("Welcome to small level calculator")
    n1 = int(input("Enter number 1:"))
    n2 = int(input("Enter number 2:"))
    print("1. addition /n 2. subtraction /n 3. multiple /n 4. division")
    choice = int(input("Enter your choice:"))
    if choice == 1:
        print("addition of two numbers s:", addition(n1,n2))
    elif choice == 2:
        print("subtraction of two numbers s:", subtraction(n1,n2))
    elif choice == 3:
        print("multiple of two numbers s:", multiple(n1,n2))
    elif choice == 4:
        print("division of two numbers s:", division(n1,n2))
    else :
        print("please select correct operation")
```
