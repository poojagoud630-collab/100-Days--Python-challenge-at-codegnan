```python
# square number pattern
n=4
k=1
for i in range(1,n+1):
    for j in range(1,n+1): 
        print(k, end = " ")
        k+=1
    print()
```

```python
# right angled triangle number pattern
n=4
k=1
for i in range(1,n+1):
    for j in range(1,n+1): 
        if i>=j:
          print(k, end = " ")
          k+=1
        else:
            print( " ",end=" ")
    print()
```

```python
# hallow right angle triangle 
for i in range(6):
    for j in range(6):
        if i == j or j==0 or i==n-1 :
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()  
```

```python
# Reverse hallow right angle triangle 
for i in range(6):
    for j in range(6):
        if i == j or i==0 or j==n-1 :
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print() 
```
